```python
# Importing numpy as np
import numpy as np
x = np.array([10,12,14,16,17])
y = np.array([20,22,24,26,27])
x, y
```

```python
x = tf.constant(x)
y = tf.constant(y)
x, y
```

```python
# Finding the maximum and minimum values in x & y
tf.reduce_max(x),
tf.reduce_min(y)
```

```python
x = tf.cast(x, dtype = tf.float32)
y= tf.cast(y, dtype = tf.float32)
```

```python
# 1. Create a model
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(100, name = "Input_layer"),
                             #tf.keras.layers.Dense(10),
                             tf.keras.layers.Dense(1, name = "Output_layer"),


], name = "model_1")

# 2. Compile the model
model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01),
              metrics = ["mae"]
)
```

```python
# Visualizing the summary
model.summary()
```

```python
# fitting the model
model.fit(tf.expand_dims(x, axis = -1), y, epochs =100)
```

```python
model.predict([29])
```
