# Dot Product

```python
np.random.seed(0)

mat1 = np.random.randint(10, size =(5,3))
mat2 = np.random.randint(10, size =(5,3))

mat1
```

```python
mat2
```

```python
mat1.shape, mat2.shape
```

```python
# Element wise matrix multiplication (Hardmard product)
mat1 * mat2
```

```python
# Dot Porduct
np.dot(mat1, mat2)
mat1.T.shape
```

```python
mat2.T.shape
```

```python
# For dot product number of columns of 1st matrix should be equal to the number of rows of 2nd
mat3 = np.dot(mat1, mat2.T)
mat3
```

# Dot prodeuct example(nut butter sales)

```python
np.random.seed(0)
#Number of jars sold
sales_amounts = np.random.randint(20, size= (5,3))
sales_amounts
```

```python
# Create weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts,
                            index = ["Mon", "Tues", "Wed", "Thurs", "Fri"],
                            columns = ["Almond butter", "Peanut butter", "Cachew butter"])
weekly_sales
```

```python
#Create prices array
prices = np.array([10, 8, 12])
prices
```

```python
#Create butter _prices DataFrame
butter_prices = pd.DataFrame(prices.reshape(1,3),
                             index = ["Price"],
                             columns = ["Almond butter", "Peanut butter", "Cachew butter"])
```

```python
butter_prices
```

```python
# Shape aren't aligned, let's transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
# Create daily sales
butter_prices.shape, weekly_sales.shape
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
# Doesn't work not the right shape
weekly_sales["Total ($)"] = daily_sales
weekly_sales
```

```python
weekly_sales["Total ($)"]= daily_sales.T
weekly_sales
```
